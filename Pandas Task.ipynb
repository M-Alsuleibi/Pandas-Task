```python
#Drop the following columns: new_deaths_smoothed','new_cases_per_million','total_cases_per_million’
df_copy=df.copy()
df.drop(['new_deaths_smoothed','new_cases_per_million','total_cases_per_million'],axis=1,inplace=True)
```

```python
#Rename the following columns:'date': 'Date','location':'Country','continent': 'Continent','iso_code':'ISO_code’
df.rename(columns={'date': 'Date','location':'Country','continent': 'Continent','iso_code':'ISO_code' },inplace=True )
```

```python
#List the continent name: Create a list that holds the unique values of continent name
continent_unique=list(df.Continent.unique)
```

```python
#There is something in sklearn library called Simple imputer helps with missing values in a dataset,
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
df2
```

```python
#Groupby, df.groupby() : The groupby command allows us to divide our data into different groups and perform our data analysis.
df3 = df2.groupby(['Date','Country',])[['total_cases','total_deaths','total_vaccinations']].sum().reset_index()
df3
```

```python
#Replace “missing_value” to 0 for the following columns: [’total_cases’,’total_deaths’,’total_vaccinations’]
df2['total_cases'] = df2['total_cases'].replace(['missing_value'], ['0'])
df2['total_deaths'] = df2['total_deaths'].replace(['missing_value'], ['0'])
df2['total_vaccinations'] = df2['total_vaccinations'].replace(['missing_value'], ['0'])
df2
```

```python
#From df3: Find total countries where total_deaths is greater than 1000000.
df3['total_deaths'] = pd.to_numeric(df3['total_deaths'], errors='coerce')

total_countries = len(df3[df3['total_deaths'] > 1000000])
print(total_countries)
```

```python
#How many dates we have in total where total_deaths is greater than 1000000
total_dates = len(df3[df3['total_deaths'] > 1000000]['Date'].unique())
print(total_dates)
```
